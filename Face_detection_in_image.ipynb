```python
face_cascade = cv2.CascadeClassifier('/content/gdrive/My Drive/haarcascade_frontalface_default.xml') # We load the cascade for the face.
eye_cascade = cv2.CascadeClassifier('/content/gdrive/My Drive/haarcascade_eye.xml') # We load the cascade for the eyes.
```

```python
img = cv2.imread('/content/gdrive/My Drive/FuLIJk6aAAIGQIi.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # We do some colour transformations.
```

```python
faces = face_cascade.detectMultiScale(gray, 1.3, 5)# We apply the detectMultiScale method from the face cascade to locate one or several faces in the image.
```

```python
for (x,y,w,h) in faces: # For each detected face:
     cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)# We paint a rectangle around the face.
     roi_gray = gray[y:y+h, x:x+w] # We get the region of interest in the black and white image.
     roi_color = img[y:y+h, x:x+w] # We get the region of interest in the colored image.
     eyes = eye_cascade.detectMultiScale(roi_gray) # We apply the detectMultiScale method to locate one or several eyes in the image.
     for (ex,ey,ew,eh) in eyes: # For each detected eye:
         cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2) # We paint a rectangle around the eyes, but inside the referential of the face.
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```
